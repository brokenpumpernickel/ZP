## Named Tuples

```python
from typing import NamedTuple
```

```python
class Student(NamedTuple):
    name: str
    age: int
    email: str
    classes: list[str]
```

```python
bob = Student(name='Bob',
              age=21,
              email='bob@bulszit.pl',
              classes=['Math', 'Physics'])
```

```python
bob
```

```python
bob.name, bob.age
```

```python
bob.emil
```

```python
bob[0], bob[1], bob[2], bob[3]
```

```python
bob['name']
```

```python
bob.name = 'Robert'
```

## Dataclasses

```python
from dataclasses import dataclass, field
```

```python
@dataclass
class Student:
    name: str
    age: int
    email: str
```

```python
bob = Student(name='Bob', age=21, email='bob@bulszit.pl')
```

```python
bob.name
```

```python
bob.age = 60
```

```python
bob.emil = 'bob@shitebul.pl'
```

```python
bob.email
```

```python
bob
```

```python
@dataclass(slots=True)
class Student:
    name: str
    age: int
    email: str
```

```python
bob = Student(name='Bob', age=21, email='bob@bulszit.pl')
```

```python
bob.emil = 'bob@szitbul.pl'
```

```python
@dataclass
class Student:
    name: str
    age: int = field(default_factory=lambda: 20)
    email: str = ''
    #classes: list[str]
```

```python
bob = Student(name='Bob')
```

```python
bob
```

```python
@dataclass
class Student:
    name: str
    age: int
    email: str
    classes: list[str] = field(default_factory=list)

    def __post_init__(self):
        if '@' not in self.email:
            raise ValueError('Wpisałeś zły mail idioto!')
```
