```python
def my_generator():
    yield "Hello, World!"
    yield "Welcome to Python generators."
    yield "Generators are memory efficient."
    yield "You can iterate over them just like lists."
    yield "That's all for now!"
```

```python
my_generator()
```

```python
for val in my_generator():
    print(val)
```

```python
def my_range(n):
    i = 0
    while i < n:
        yield i
        i += 1
```

```python
for i in my_range(5):
    print(i)
```

```python
class Container:
    def __init__(self):
        self.items = ['apple', 'banana', 'cherry']
    
    def __iter__(self):
        for item in self.items:
            yield item

```

```python
x = Container()

for item in x:
    print(item)
```
